```python
text = str()
with open("../dumps-text/fr_total2.txt", "r") as f:
    text = f.read()
#text
```

```python
import string
import re

text = text.lower()
#text
spec_chars = string.punctuation + '\n\xa0«»\t—…'
text = re.sub('\n', '', text)
#text
text = "".join([ch for ch in text if ch not in spec_chars])
#text
```

```python
from nltk.corpus import stopwords
from nltk import word_tokenize
import nltk

text_tokens = word_tokenize(text)
text = nltk.Text(text_tokens)

french_stopwords = stopwords.words("french")
french_stopwords.extend(['http', 'comme', 'donc', 'dont', 'quand', 'direct', 'ca', 'plus', 'button', 'sousmenu', 'node', 'https', 'france bleu', 'france bleu ', 'gerald darmanin', 'html', 'post2020', 'rfi', 'type', 'file', 'localhost', 'dy', 'daccueil', 'ajouter', 'com', 'cest', 'ça', 'si', 'min', 'très', 'san'])

text_tokens = [token.strip() for token in text_tokens if token not in french_stopwords]
text = nltk.Text(text_tokens)
print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

text_raw = " ".join(text)

wordcloud = WordCloud().generate(text_raw)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
